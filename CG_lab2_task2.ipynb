```python
from skimage.io import imshow, imsave, imread
from skimage import img_as_float
from skimage.color import rgb2gray, rgb2hsv, hsv2rgb
from numpy import dstack
import numpy as np
from matplotlib import pyplot as plt
```

```python
def hist_red(path):
    img = imread(path)
    plt.hist(img[:,:,0], facecolor = "red" );
    plt.xlabel("Intensity");
    plt.ylabel("Count of pixels");

def hist_green(path):
    img = imread(path)
    plt.hist(img[:,:,1], facecolor = "green");
    plt.xlabel("Intensity");
    plt.ylabel("Count of pixels");

def hist_blue(path):
    img = imread(path)
    plt.hist(img[:,:,2], facecolor = "blue");
    plt.xlabel("Intensity");
    plt.ylabel("Count of pixels");
    
def show_splitted_rgb(path):
    img = imread(path)
    red_channel = img.copy()
    blue_channel = img.copy()
    green_channel = img.copy()
    
    red_channel[:, :, 1] = 0
    red_channel[:, :, 2] = 0
    
    green_channel[:, :, 0] = 0
    green_channel[:, :, 2] = 0
    
    blue_channel[:, :, 0] = 0
    blue_channel[:, :, 1] = 0
    
    fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(10, 10))
    
    ax[0, 0].imshow(img)
    ax[0, 0].set_title("Original")
    
    ax[0, 1].imshow(red_channel)
    ax[0, 1].set_title("Red")
    
    ax[1, 0].imshow(green_channel)
    ax[1, 0].set_title("Green")
    
    ax[1, 1].imshow(blue_channel)
    ax[1, 1].set_title("Blue")
```
